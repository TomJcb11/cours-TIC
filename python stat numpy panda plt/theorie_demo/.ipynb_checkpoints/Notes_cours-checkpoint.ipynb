```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N = 5000

math_males_avg = 0.7
math_females_avg = 0.5
eng_males_avg = 0.5
eng_females_avg = 0.7

std_males = 0.1
std_females = 0.08

mean = [0,0]
cov = [[1 , 0.5]
      ,[0.5, 1 ]]

Z_math, Z_eng = np.random.multivariate_normal(mean, cov, N).T

math_m = np.clip(Z_math*std_males + math_males_avg,0,1)
eng_m = np.clip(Z_eng*std_males + eng_males_avg,0,1)

math_f = np.clip(Z_math*std_females + math_females_avg,0,1)
eng_f = np.clip(Z_eng*std_females + eng_females_avg,0,1)

males = pd.DataFrame({"math":math_m,"eng":eng_m})
males["gender"] = 'M'
females = pd.DataFrame({"math":math_f,"eng":eng_f})
females["gender"] = "F"

all = pd.concat([males,females],axis=0)
all["class"] = np.random.choice(["A","B","C"],2*N)
```

```python
plt.scatter(math_f, eng_f)
plt.scatter(math_m, eng_m)
plt.axis([0,1,0,1])
plt.show()
```

```python
all[["math","eng"]] *= 20
all[["math","eng"]] = all[["math","eng"]].round(1)

all.to_csv("demo/d1_scores.csv",index=False,mode='w')
```
