## Test d'hypotheses

### le but, l'intérêt, l'idée générale

Admettons que:
- on ait 2 echantillons $X,Y$ d'une même population $A$;
- on ait 1 echantillon $Z$ d'une autre population $B$;
- ces populations suivent une distribution statistique similaire.
- pour l'illustration, on ne sait pas quelles echantillons viennent de quelle population


Conséquences:
- pour chaque echantillon, il peut y avoir des fluctuation aléatoires par apport à leur population respectives;
- on observera (presque) toujours des différences entre différents echantillon d'une même population.
- sur bases de ces différences naivement observé, on ne peut pas dire si les 3 echantillons viennent d'une même population ou non.

La question à ce poser ce n'est donc jamais si il y a des différences (= oui), mais plutot si les différences observées sont "statistiquement significatives" et pas juste à cause de fluctuation statistiques (=coinsidences).

Les tests d'hypotheses on pour but de ***quantifier la probabilité*** que des echantillons serait issu d'une ***même population***, sur base des observations.

## plusieurs tests possibles:


### distribution discrètes
- test de $\chi^2$ (chi carré) : dans un tableau de contengence, est-ce qu'il y a une différence significative entre les effectifs observés et les effectif théoriques


### distribution normale
- test de normalité (Shapiro-Wilk) : est-ce que la distribution suit une Gaussienne/normale
- test sur la moyenne (t-test, Welch) : est-ce que les moyennes de deux distributions normales sont égales
- test sur la variance (F-fisher, Levene) : est-ce que les variances de deux distributions normales sont égales

### pleins d'autres tests
 - ...


## les hypotheses nulles de tests

Par construction, chaque test ne permet de dire qu'une seule chose bien présise: quel est la probabilité d'avoir les résulats obtenus, si on ce trouve dans le scénario de l'hypothese nulle $H_0$.


Prenons comme exemple le test sur les moyennes:
- on souhaite savoir si les moyennes observés sont statistiquement différentes.  
- l'hypotese nulle $H_0$ du test (Student T-test)
    - (les expressions suivantes sont équivalentes):
        1. les moyennes observés $\bar{x}_1, \bar{x}_2$ viennent d'échantillons d'une même population
        2. les moyennes sont statistiquement identiques
        3. les moyennes sont, avec quelques fluctuation statistiques, globalement les mêmes

L'hypothese nulle (= spécifique au test) ne correspond pas à ce qu'on souhaite savoir.
Cependant, en suivant la procédure de conclusion des test d'hypotheses (voir plus bas), je vais pouvoir arriver à conclure si se que je souhaite démontrer est (statistiquement) vrai ou non.



## Pourquoi ces hypotheses nulles-là ?

Il faut savoir que chaque test est défini comme un calcul ou une séries de calculs.

Dans la construction de ce calcul, pour que le calcul ait un sens, on a besoin de poser des hypotheses de départ: l'hypothese nulle $H_0$.

(On appelle le résultat du calcul: "la statstique du test")


## Le résulat d'un test:

Ce qu'on souhaite obtenir d'un test, c'est la probabilité que l'hypothese nulle est vraie.  
Cette probabilité, c'est la "p-valeur"  
(EN:"p-value").

Pour chaque test et pour chaque résultat du calcul du test (statstique du test), une "p-value" lui est associé .

Exemple: 
1. la "p-value" est élevé (*):
    - on va admettre que l'hypothese nulle est probablement vraie.

2. la "p-value" est faible (*):
    - on va admettre que l'hypothèse nulle est probablement fausse.
    - Par conséquence, on va accepter l'hypothese altenative $H_a$ , qui la formulation (logiquement) opposé à $H_0$
        - si $H_0: \bar{x}_1 = \bar{x}_2$, alors $H_a: \bar{x}_1 \neq \bar{x}_2$
        - si $H_0: \bar{x} = 0$, alors $H_a: \bar{x} \neq 0$
        - si $H_0:$ "à gauche", alors $H_a:$ (~~à droite~~) tout, sauf "gauche"

(*): on doit encore détailler quel est le seuil à utiliser pour tirer nos conclusions.



## les types d'erreurs

Il faut noter que comme la "p-value" est une probabilité (et $p > 0$), on ne peut mathématiquement jamais exclure que l'hypothese nulle $H_0$ est vraie ou fausse.

### risque type 1:

Quand on veut conclure que l'hypothese $H_0$ est fausse parce que la "p-value" est faible (*) , il y a une probabilité de se tromper. On appelle ce risque le "risque de type 1".

Exemple:
- $p=0.03$: la probabilité que $H_0$ est vraie, est de 3%. Si je décide de conclure que $H_0$ est fausse, j'ai donc "3% de chance" de me tromper en choissant cette conclusion.


### risque type 2:
Quand on va conclure que l'hypothese $H_0$ est vraie parce que la "p-value" est élevé (*), il y a également une probabilité de se tromper de conclusion. On appelle ce risque le "risque de type 2".

Cela ce produit essensiellement quand on veut mesurer une différence entre des phénomènes :  
- pour lesquelles on a pas assez de données pour établir des différences
- pour lesquelles les différences sont très subtils, et donc nécéssite plus de données pour mettre en evidence les différences.

En bref, ce risque est présent parce qu'on n'a (potentiellement) pas assez de données.

En conclusion, si la "p-value" est élevé (*), il n'y a pas de différences statistiquement significatifs, car:
- les distribution sont identiques  
ou  
- il n'y a pas suffissament d'élément/données pour confirmer cette différence.

## la procedure de conclusion de test d'hypotheses:

1. Choisir le test d'hypothese à effectuer
2. Voir quel est l'hypothese $H_0$ lié au test
3. **Decider à l'avance** quel est niveau de **risque toléré** pour le "risque de type 1": $\alpha$
4. evaluer la "p-value":
    - $p<\alpha$ : rejeter l'hypothèse $H_0$, accepter $H_a$ comme vraie (d'après les observations)
    - $p\geq\alpha$: garder l'hypothèse $H_0$, accepter $H_0$ comme vraie (d'après les observations)


en mots:  
- si je rejette l'hypothese $H_0$, la probabilité de me tromper est de $p$ 
    - si $p$ est plus faible que le seuil toléré $\alpha$
        - alors je rejette $H_0$, je prends $H_a$ comme vraie
        - la conclusion sur les echantillons :
            - il y a une différence trop importante (selon le seuil établi) entre les echantillons pour que cela soit dû au hasard
                 - les différences ne sont pas dû à une coïnsidence
            - il y a malgré tout quand-même une probabilité faible que cela soit dû au hasard 
    - si $p$ est plus élévé que le seuil toléré $\alpha$
        - alors je ne peut pas me permettre de rejetter $H_0$, je prends/garde $H_0$ comme vraie
        - la conclusion sur les echantillons :
            - si les echantillons devait venir de source différentes, je n'ai pas suffisamment d'éléments pour pouvoir établir cette différence
                - les différences sont dû à une coïnsidence
            - si les echantillons devait venir de la même source, les données me conforte dans cette hypothese

## Demo

nous allons prendre un seuil de tolérence $\alpha = 0.05$ pour tous les tests:
- cela sigifie que pour chaque conclusion, on se permet d'avoir une erreur de type 1 (au maximum) "1 fois sur 20" 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
### construction du dataset
N_petit = 10
X_petit = np.random.normal(179,6.5,(N_petit)) # taille homme
Y_petit = np.random.normal(179,6.5,(N_petit)) # taille homme
Z_petit = np.random.normal(164,5.5,(N_petit)) # taille femme
V_petit = 0.5*(np.random.uniform(150,200, (N_petit)) + np.random.uniform(150,200, (N_petit)))

N = 2000
X = np.random.normal(179,6.5,(N)) # taille homme
Y = np.random.normal(179,6.5,(N)) # taille homme
Z = np.random.normal(164,5.5,(N)) # taille femme
V = 0.5*(np.random.uniform(150,200, (N)) + np.random.uniform(150,200, (N)))
```

```python
petit_dataset = {"X":X_petit, "Y":Y_petit, "Z":Z_petit, "V":V_petit}
grand_dataset = {"X":X, "Y":Y, "Z":Z, "V":V}
```

```python
pars = {'alpha':0.5,"bins":range(150,210,4)}

plt.hist(V_petit,**pars, color = 'b', label = 'homme')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.title('histogrammes de petits datasets')
plt.hist(X_petit,**pars, color = 'b', density=True,label="X") 
plt.hist(Y_petit,**pars, color = 'g', density=True,label="Y" )
plt.hist(Z_petit,**pars, color = "r", density=True,label="Z")
#plt.hist(V_petit,**pars, color = "cyan", density=True )
plt.legend()

plt.subplot(1,2,2)
plt.title('histogrammes de grands datasets')
plt.hist(X,**pars, color = 'b', density=True,label="X") 
plt.hist(Y,**pars, color = 'g', density=True,label="Y")
plt.hist(Z,**pars, color = "r", density=True,label="Z")
#plt.hist(V,**pars, color = "cyan", density=True )
plt.legend()

plt.show()
```

## [Shapiro-Wilk : Normalité](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html)
"""  
Perform the Shapiro-Wilk test for normality.  
The Shapiro-Wilk test tests the null hypothesis that the data was drawn from a normal distribution.  
"""  
on peut également utiliser [un autre test](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.normaltest.html#scipy.stats.normaltest) de normalité

```python
print("petit_dataset")
for key,val in petit_dataset.items():
    print(f'dataset = {key}, resultat = {stats.shapiro(val)}')

print("-"*50)
print("grand_dataset")
for key,val in grand_dataset.items():
    print(f'dataset = {key}, resultat = {stats.shapiro(val)}')
```

**CONCLUSIONS**:
- rappel
    - Hypothese nulle $H_0$ = distribution normale
    - Hypothese altérnative $H_a$ = pas une distribution normale
<br><br>
- "grand_dataset\[v\]" : $p<alpha$
    - donc on rejette l'hypothese nulle $H_0$, prend $H_a$ comme vraie
        - donc pas une distribution normale

- pour les autres données, je n'ai pas pu constater des différences d'une par apport à une distibution normale
    - et cela, même pour V_petit

## T-test pour comparer les moyennes:
il faut que les distributions utilisés soient (approximativement) normales.
### indépendants:

si les données sont des echantillons indépendants: **[celui-ci](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html#scipy.stats.ttest_ind)**
    - les individus des deux echantillons sont différents

"""  
Calculate the T-test for the means of two independent samples of scores.

This is a test for the null hypothesis that 2 independent samples have identical average (expected) values. This test assumes that the populations have identical variances by default.  
"""  

### dépendants:

si les données sont des echantillons dépendants: **[celui-là](https://docs.scipy.org/doc/scipy-1.14.0/reference/generated/scipy.stats.ttest_rel.html)**
    - ce sont les mêmes individus, mais des mesures différentes (par exemple: à une date différente)

"""  
Calculate the t-test on TWO RELATED samples of scores, a and b.

This is a test for the null hypothesis that two related or repeated samples have identical average (expected) values.  
"""  


```python
### ici les données sont des données indépendantes

print("petit_dataset")
combinaisons = [("X","Y"),("X","Z"),("X","V")]

for a,b in combinaisons:
    print(f't-test entre {a} et {b}, resultat= {stats.ttest_ind(petit_dataset[a],petit_dataset[b])}')

print("-"*50)
print("grand_dataset")
for a,b in combinaisons:
    if (a!="V") and (b!="V"): ### V, n'est plus éligible, car ne suit pas une normale
        print(f't-test entre {a} et {b}, resultat= {stats.ttest_ind(grand_dataset[a],grand_dataset[b])}')
```

**CONCLUSIONS**:
- rappel
    - $H_0$ : les moyennes sont identiques
<br><br> 
- petit_dataset:
    - les tests nous suggèrent que les moyennes sont (statistiquement) différentes (pour toutes les combaisons)
    - sauf pour la combinaison (X,Z)
- grand_dataset:
    - pour (X,Z): les moyennes sont (statistiquement) différentes

### ANOVA/F_test :  comparer >2 dataset Gaussien : moyennes

```python
plt.boxplot([X,Y,Z])
plt.show()
```

```python
print(stats.f_oneway(X,Y,Z))
```

```python
print(stats.f_oneway(X,Y))
print(stats.ttest_ind(X,Y)) ## +-= p-value précédent
```

## [comparaison des variances : Levene](https://docs.scipy.org/doc/scipy-1.14.0/reference/generated/scipy.stats.levene.html)
"""  
Perform Levene test for equal variances.

The Levene test tests the null hypothesis that all input samples are from populations with equal variances. Levene’s test is an alternative to Bartlett’s test bartlett in the case where there are significant deviations from normality.
  
"""  

```python
print("petit_dataset")
combinaisons = [("X","Y"),("X","Z"),("X","V")]

for a,b in combinaisons:
    print(f't-test entre {a} et {b}, resultat= {stats.levene(petit_dataset[a],petit_dataset[b])}')

print("-"*50)
print("grand_dataset")
for a,b in combinaisons:
    if (a!="V") and (b!="V"): ### V, n'est plus éligible, car ne suit pas une normale
        print(f't-test entre {a} et {b}, resultat= {stats.levene(grand_dataset[a],grand_dataset[b])}')
```

## [Correlations](https://docs.scipy.org/doc/scipy-1.14.0/reference/generated/scipy.stats.pearsonr.html)
"""  
Pearson correlation coefficient and p-value for testing non-correlation.

The Pearson correlation coefficient [1] measures the linear relationship between two datasets. Like other correlation coefficients, this one varies between -1 and +1 with 0 implying no correlation. Correlations of -1 or +1 imply an exact linear relationship. Positive correlations imply that as x increases, so does y. Negative correlations imply that as x increases, y decreases.

This function also performs a test of the null hypothesis that the distributions underlying the samples are uncorrelated and normally distributed. (See Kowalski [3] for a discussion of the effects of non-normality of the input on the distribution of the correlation coefficient.) The p-value roughly indicates the probability of an uncorrelated system producing datasets that have a Pearson correlation at least as extreme as the one computed from these datasets.

"""  

```python
plt.scatter(X,Y)
plt.show()
stats.pearsonr(X,Y) # H0 = uncorrelated
```

**CONCLUSIONS**:
- ongarde l'hypothese nulle
- la correlation n'est pas significative (= les différences sont du à une coïnsidence)
- et le degrée/coéficient de corrélation est **exactement 0**

```python
X_bis  = X+ 10*np.random.normal(0,1,len(X))

plt.scatter(X,X_bis)
plt.show()

stats.pearsonr(X,X_bis)
```

**CONCLUSIONS**:
- on rejete l'hypothese nulle
- la correlation est significative (=pas une coïnsidence)
- et le degrée/coéficient de corrélation est 0.54..

```python
X_tris  = X+ 3*np.random.normal(0,1,len(X))

plt.scatter(X,X_tris)
plt.show()

stats.pearsonr(X,X_tris)
```

# [Chi2: comparer des effectifs](https://docs.scipy.org/doc/scipy-1.14.0/reference/generated/scipy.stats.chisquare.html)

"""  
Calculate a one-way chi-square test.

The chi-square test tests the null hypothesis that the categorical data has the given frequencies.  
"""  



```python
a_expected = np.array([7,3,5,10]) # nombre occurences pour cat = A,B,C,D
a_observed = np.array([6,4,4,11])

stats.chisquare(f_obs = a_observed,f_exp = a_expected) # H0: the same (theoretical) distribution
```

```python
b_expected = np.array([7,3,5,10])
b_observed = np.array([10,10,5,0])

stats.chisquare(f_obs = b_observed,f_exp = b_expected)
```

```python
r, pval = stats.chisquare(f_obs = b_observed,f_exp = b_expected) 

print(f"statistic: {r}, pval:{pval} ")
```
