# Différence is / == 

pour des entiers ca marche parce qu'il n'existe qu'une seule classe entier poour chacune des valeurs . P/ex toutes les variables qui contiennent le int **12** vont **pointer** vers la classe **12** unique. 

!! selon le mode d'exécution de python (par fichier de façon traditionnelle ou par ligne dans le cadre de la console/notebook) l'allocation en mémoire se déroule de façon différente 





```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]
d = int(300)
e = int(300)

#print("a is b => ",a is b)  # True, car a et b pointent vers le même objet en mémoire
#print("a is c => ",a is c)  # False, car a et c sont des objets différents en mémoire
print("d is e =>",d is e)  # True, car d et e pointent vers le même objet en mémoire (pour les entiers de -5 à 256)
#print("a == c =>",a ==c)  # True, car a et c ont les mêmes valeurs
#print("d == e =>",d==e)  # True, car d et e ont les mêmes valeurs
print(id(d))
print(id(e))
```
