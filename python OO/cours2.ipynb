```python
list = [1, 2, 3, 33, 4, 5, 1, 2, 5, 4, 5]
list[-1] = 6

list.append(5)

list.insert(4, 3)

list.remove(3)

list.pop(4)

list2 = list[3:]

print(list, id(list))
print(list2, id(list2))
print(id(list) == id(list2))
```

# Distinction dictionnaire Python et JSON

| Aspect       | Dictionnaire Python                           | JSON                                        |
| ------------ | --------------------------------------------- | ------------------------------------------- |
| Délimitation | `{}`                                          | `{}`                                        |
| Clés         | N'importe quel type immuable                  | Chaînes de caractères                       |
| Valeurs      | N'importe quel type                           | Types JSON valides (chaînes, nombres, etc.) |
| Utilisation  | Stockage et manipulation de données en Python | Échange de données entre serveur et client  |
| Conversion   | Utilise le module `json` pour conversion      | Utilise le module `json` pour conversion    |
